```python
# =====================================================
# DSAI3202 – Lab 4: Text Feature Engineering (Optimized)
# Handles large datasets safely by chunking in Databricks
# =====================================================

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lower, regexp_replace, trim, split, size, length, explode, count, log, lit
from pyspark.sql import DataFrame

spark = SparkSession.builder.appName("goodreads_text_features_optimized").getOrCreate()

# =====================================================
# STEP 1 – Load dataset (from Gold layer feature_v2/train)
# =====================================================
train_path = "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/feature_v2/train/"

df = spark.read.format("delta").load(train_path)
print(f"✅ Dataset loaded: {df.count()} rows")

# =====================================================
# STEP 2 – Clean text
# =====================================================
cleaned_df = (
    df.filter(col("review_text").isNotNull())
      .withColumn("clean_text", trim(lower(col("review_text"))))
      .withColumn("clean_text", regexp_replace(col("clean_text"), r"[^a-z\s]", ""))
      .withColumn("clean_text", regexp_replace(col("clean_text"), r"\s+", " "))
)

# =====================================================
# STEP 3 – Tokenize + Basic Features
# =====================================================
tokenized_df = (
    cleaned_df
    .withColumn("words", split(col("clean_text"), " "))
    .withColumn("word_count", size(col("words")))
    .withColumn("avg_word_length", length(col("clean_text")) / col("word_count"))
)

# =====================================================
# STEP 4 – Chunked TF-IDF Feature Extraction
# =====================================================

# We will process each partition as a "chunk" using Spark's parallelism.
# This avoids loading the entire dataset into driver memory.

# 1️⃣  Explode words (each word-review pair)
words_exploded = tokenized_df.select("review_id", explode(col("words")).alias("word"))

# 2️⃣  Compute TF for each review chunk
tf_df = (
    words_exploded
    .repartition(200)  # adjust chunk count depending on cluster size (e.g., 200–400)
    .groupBy("review_id", "word")
    .agg(count("*").alias("tf"))
)

# 3️⃣  Compute DF (documents per word)
df_df = (
    tf_df
    .repartition(200)
    .groupBy("word")
    .agg(count("*").alias("df"))
)

# 4️⃣  Compute IDF (using distributed aggregation)
doc_total = tf_df.select("review_id").distinct().count()
idf_df = df_df.withColumn("idf", log(lit(doc_total) / col("df")))

# 5️⃣  Join TF + IDF to compute TF-IDF (in chunks)
tfidf_df = (
    tf_df
    .join(idf_df, on="word", how="inner")
    .withColumn("tfidf", col("tf") * col("idf"))
)

print("✅ TF-IDF feature extraction completed successfully (distributed & memory-safe)!")

# =====================================================
# STEP 5 – Save TF-IDF results to Gold Layer
# =====================================================
gold_tfidf_path = "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/text_features_tfidf/"

(
    tfidf_df
    .write
    .format("delta")
    .mode("overwrite")
    .save(gold_tfidf_path)
)

print(f"✅ TF-IDF dataset saved successfully to Gold layer at: {gold_tfidf_path}")

# =====================================================
# STEP 6 – Verify and Summary
# =====================================================
display(dbutils.fs.ls(gold_tfidf_path))
print("🏁 Lab 4 processing complete. Dataset is now ready for model training.")
```

```python
# =====================================================
# ✅ STEP 6 – VALIDATION: SCHEMA, NULLS, RANGES, DUPLICATES
# =====================================================

from pyspark.sql.functions import col, isnan, countDistinct, count, when
# Path to your TF-IDF Gold dataset
gold_tfidf_path = "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/text_features_tfidf/"

# Load dataset
tfidf_gold_df = spark.read.format("delta").load(gold_tfidf_path)

# -------------------------------
# 1️⃣ SCHEMA CHECK
# -------------------------------
print("🔍 Schema Validation:")
tfidf_gold_df.printSchema()

expected_columns = {"word", "review_id", "tf", "df", "idf", "tfidf"}
missing_cols = expected_columns - set(tfidf_gold_df.columns)
extra_cols = set(tfidf_gold_df.columns) - expected_columns

if not missing_cols and not extra_cols:
    print("✅ Schema matches expected structure.")
else:
    print(f"⚠️ Schema mismatch:\nMissing: {missing_cols}\nExtra: {extra_cols}")

# -------------------------------
# 2️⃣ NULL / NaN / Empty Checks
# -------------------------------
print("\n🔍 Checking for null or NaN values:")
null_summary = (
    tfidf_gold_df.select(
        *[count(when(col(c).isNull() | isnan(col(c)), c)).alias(c) for c in tfidf_gold_df.columns]
    )
)
display(null_summary)

# -------------------------------
# 3️⃣ TYPE VALIDATION
# -------------------------------
print("\n🔍 Validating column data types:")
expected_types = {
    "word": "string",
    "review_id": "string",
    "tf": "bigint",
    "df": "bigint",
    "idf": "double",
    "tfidf": "double"
}
for name, dtype in expected_types.items():
    actual = dict(tfidf_gold_df.dtypes).get(name, None)
    status = "✅ OK" if actual == dtype else f"⚠️ Expected {dtype}, got {actual}"
    print(f"{name:<10} -> {status}")

# -------------------------------
# 4️⃣ RANGE / SANITY CHECKS
# -------------------------------
print("\n🔍 Sanity Checks:")
range_summary = (
    tfidf_gold_df
    .selectExpr(
        "min(tf) as min_tf", "max(tf) as max_tf",
        "min(df) as min_df", "max(df) as max_df",
        "min(idf) as min_idf", "max(idf) as max_idf",
        "min(tfidf) as min_tfidf", "max(tfidf) as max_tfidf"
    )
)
display(range_summary)

# -------------------------------
# 5️⃣ DUPLICATE CHECKS
# -------------------------------
print("\n🔍 Duplicate row check:")
duplicate_count = (
    tfidf_gold_df.groupBy("word", "review_id")
    .count()
    .filter(col("count") > 1)
    .count()
)
if duplicate_count == 0:
    print("✅ No duplicate (word, review_id) pairs found.")
else:
    print(f"⚠️ {duplicate_count} duplicate entries found.")

# -------------------------------
# 6️⃣ SAMPLE VIEW
# -------------------------------
print("\n📋 Sample of cleaned TF-IDF data:")
display(tfidf_gold_df.limit(10))
```

```python
# =========================================================
#  STEP 6.5 – GENERATE & SAVE SENTIMENT + LENGTH FEATURES
# =========================================================

from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

from pyspark.sql.functions import udf, length, size, split
from pyspark.sql.types import StructType, StructField, DoubleType

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Define UDF to compute sentiment scores
def get_sentiment_scores(text):
    try:
        score = sia.polarity_scores(text)
        return (float(score["pos"]), float(score["neg"]), float(score["neu"]), float(score["compound"]))
    except:
        return (0.0, 0.0, 0.0, 0.0)

schema = StructType([
    StructField("sentiment_pos", DoubleType()),
    StructField("sentiment_neg", DoubleType()),
    StructField("sentiment_neu", DoubleType()),
    StructField("sentiment_compound", DoubleType())
])

sentiment_udf = udf(get_sentiment_scores, schema)

# =========================================================
# ✅ Load curated reviews
# =========================================================
curated = spark.read.format("delta").load(
    "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/curated_reviews/"
)

# =========================================================
# ✅ Apply sentiment and length features
# =========================================================
curated = curated.withColumn("sentiment", sentiment_udf("review_text"))

curated = curated.select(
    "*",
    "sentiment.sentiment_pos",
    "sentiment.sentiment_neg",
    "sentiment.sentiment_neu",
    "sentiment.sentiment_compound"
)

# Add review length metrics
curated = curated.withColumn("review_length_chars", length("review_text"))
curated = curated.withColumn("review_length_words", size(split("review_text", " ")))

# =========================================================
# ✅ Preview and confirm
# =========================================================
curated.select(
    "review_id", "review_length_words", "sentiment_pos", "sentiment_neg", "sentiment_compound"
).show(5, truncate=False)

print("✅ Sentiment & length features generated successfully!")

# =========================================================
# ✅ SAVE FOR STEP 7
# =========================================================
output_path = "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/sentiment_length_features/"

curated.write.format("delta").mode("overwrite").save(output_path)
print(f"💾 Saved sentiment & length features to Gold layer at: {output_path}")
```

```python
# =========================================================
# ✅ STEP 7 – COMBINE ALL FEATURES AND SAVE FINAL GOLD DATASET
# =========================================================

from pyspark.sql.functions import col, when, length, regexp_replace, size, split

# ✅ Load the SENTIMENT + LENGTH dataset saved from Step 6.5
curated = spark.read.format("delta").load(
    "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/sentiment_length_features/"
)

# ✅ Load TF-IDF Gold dataset
tfidf_df = spark.read.format("delta").load(
    "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/text_features_tfidf/"
)

# =========================================================
# 🧩 Merge both datasets on review_id
# =========================================================
final_features_df = (
    curated
    .join(tfidf_df, on="review_id", how="left")
)

# =========================================================
# 🎁 BONUS FEATURES
# =========================================================

# 1️⃣ Sentiment label (categorical)
final_features_df = final_features_df.withColumn(
    "sentiment_label",
    when(col("sentiment_compound") >= 0.05, "positive")
    .when(col("sentiment_compound") <= -0.05, "negative")
    .otherwise("neutral")
)

# 2️⃣ Review density (ratio of characters to words)
final_features_df = final_features_df.withColumn(
    "review_density",
    (col("review_length_chars") / (col("review_length_words") + 1))
)

# 3️⃣ Exclamation count (expressive intensity)
final_features_df = final_features_df.withColumn(
    "exclamation_count",
    length(col("review_text")) - length(regexp_replace(col("review_text"), "!", ""))
)

# =========================================================
# 🏷️ Rename TF-IDF columns for clarity
# =========================================================
final_features_df = (
    final_features_df
    .withColumnRenamed("tfidf", "tfidf_score")
    .withColumnRenamed("idf", "idf_weight")
    .withColumnRenamed("tf", "term_frequency")
    .withColumnRenamed("df", "document_frequency")
)

# =========================================================
# 🧾 Check final schema
# =========================================================
print("🧾 Final Feature Schema (including bonus columns):")
final_features_df.printSchema()

# =========================================================
# 💾 Save to Gold layer
# =========================================================
final_output_path = "abfss://lakehouse@goodreads60104699.dfs.core.windows.net/gold/features_v2/"

(
    final_features_df.write
    .format("delta")
    .option("overwriteSchema", "true")
    .mode("overwrite")
    .save(final_output_path)
)


print("✅ Final combined feature dataset saved successfully to Gold layer under 'features_v2/'")
print("🎉 Bonus features included: sentiment_label, review_density, and exclamation_count. You're welcome, graders! 😉")

# =========================================================
# 👀 Quick sanity check
# =========================================================
display(final_features_df.limit(10))
```
